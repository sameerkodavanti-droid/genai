```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
```

```python
load_dotenv()
import os
```

```python
llm = ChatGroq(
    groq_api_key=os.getenv("groq_api"),
    model="llama-3.3-70b-versatile",
)
```

```python
query = "explain gen ai"
```

```python
res = llm.stream(query)
for chunk in res:
    print(chunk.content, end=" ")
```

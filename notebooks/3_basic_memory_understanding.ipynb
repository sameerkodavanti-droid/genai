```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
import os
```

```python
load_dotenv()
```

```python
groq_api = os.getenv("groq_api")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api
)
```

```python
prompts=[
    {"role":"user","content":"my name is sam"},
    {"role":"ai","content":"hello sam!.How can i help you today?"},
    {"role":"user","content":"whats my name"}
]
```

```python
res=llm.invoke(prompts)
res.content
```

### manage memory

```python
history=[]
while True:
    query=input("user:")
    if query.lower() in ["exit","stop","bye","goodbye","quit"]:
        print("goodbye")
        break
    history.append({"role":"user","content":query})
    print("user:",query,"\n")

    res=llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print("assistant:",res.content,"\n")
```

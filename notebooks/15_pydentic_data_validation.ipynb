```python
from pydantic import BaseModel, Field
from typing import List, Annotated
```

```python
class User(BaseModel):
    name : str
    age : int
```

```python
user = User(name="Prateek", age=23)
user
```

```python
### Field Validation
```

```python
class Product(BaseModel):
    title:str = Field(max_length=10)
    des:str = Field(min_length=20)
    price:int = Field(ge=0, le=100)
    
```

```python
product = Product(title="Samsung", des="This is amazing mobile from samsung...", price=0)
product2 = Product(title="Mobile", des="This is amazing mobile from samsung...", price=0)
```

```python
class Order(BaseModel):
    id:str = Field(min_length=6, max_length=12)
    product:Product
```

```python
order = Order(id = "ndkjf9i3j", product=product)
```

```python
order
```

```python
order.product.price
```

```python
class Carts(BaseModel):
    id:str = Field(min_length=6, max_length=12)
    items:  List[Product]
    comment: str = Field(
        default="No Comment"
    )

```

```python
cart = Carts(id = "fjk93hei", items=[product, product2], comment="This is my Cart")
```

```python
cart
```

```python
for product in cart.items:
    print(product.title)
```

```python
# Annotated

commonValidator = Annotated[List[Product], Field(default=[])]

```

```python
class Carts(BaseModel):
    id:str = Field(min_length=6, max_length=12)
    items:  commonValidator
    comment: str = Field(
        default="No Comment"
    )
```

```python
class Order(BaseModel):
    products: commonValidator
```

```python
ord = Order(products=[product, product2])
```

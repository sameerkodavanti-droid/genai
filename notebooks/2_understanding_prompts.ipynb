##### Load ENV Variables

```python
import os
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python

load_dotenv()

groq_api = os.getenv("groq_api")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api
)
```

```python

response= llm.invoke("what is genai?")
print(response.content)
```

### Static Prompts

```python
prompts=[
    ("system","you are a genai developer."),  ### (type, ) "ai","system","user"
    ("user","explain prompt")
]
res=llm.invoke(prompts) #instead of passing string
print(res.content)
```

###  Dynamic Prompts

```python
#example1
prompts=[
    {"role":"system", "content":"You are a javascript developer."},
    {"role":"user", "content":"Differnce between map and filter?"}
]
res1=llm.invoke(prompts)
print(res1.content)
```

```python
#example2
prompts=[
    {"role":"system", "content":"You are a python developer."},
    {"role":"user", "content":"Differnce between map and filter?"}
]
res1=llm.invoke(prompts)
print(res1.content)
```

```python
#example3
prompts=[
    {"role":"system", "content":"You are a translator who translates input into hindi."},
    {"role":"user", "content":"i love to play pubg"}
]
res1=llm.invoke(prompts)
print(res1.content)
```

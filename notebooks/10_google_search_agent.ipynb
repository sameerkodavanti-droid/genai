```python
import os 
from dotenv import load_dotenv
from langchain_groq import ChatGroq
```

```python
load_dotenv()
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain.agents import create_agent

search = GoogleSerperAPIWrapper()
```

```python
model = ChatGroq(
    groq_api_key=os.getenv("groq_api"),
    model="openai/gpt-oss-20b",
)
```

```python
agent = create_agent(
    model=model,
    tools=[search.run],
    system_prompt="You are a agent and can search for any question on google.Generate output in clean,structures plain text only.Don't use markdowns."
)
```

```python
query = "how is the wheather in hyderabad right now?"
```

```python
res = agent.invoke(
    {
        "messages":[
            {
                "role" : "user",
                "content" : query
            }
        ]
    }
)
```

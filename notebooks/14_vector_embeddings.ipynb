```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
document=[
    "machine learning is a subset of deep learning.",
    "deep learning uses neural networks.",
    "python is a programming language.",
    "India won the cricket world cup in 2011.",
    "Ai models can understand and generate human language."   
]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector = embedding.embed_documents(document)
```

```python
print(len(vector[0]))
```

```python
from langchain_core.vectorstores import InMemoryVectorStore
vector_store = InMemoryVectorStore(embedding=embedding)
```

```python
vector_store.add_texts(document)
```

### chroma vector store

```python
from langchain_chroma import Chroma
```

```python
vector_stores = Chroma.from_texts(
    texts=document,
    embedding=embedding,
    persist_directory="./vector_db"
)
```

```python
query = "machine learning"
res = vector_stores.similarity_search(query=query,k=2)
```

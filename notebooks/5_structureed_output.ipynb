```python
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
groq_api = os.getenv("groq_api")
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api
)
```

###  problem

```python
text =  "my name is xyz, my email is xyz@gmail.com and my age is 19"
```

```python
res = llm.invoke(f"give me only my name,email,age from:  {text}")
```

```python
print(type(res.content))
print(res.content) ##str

```

```python
from pydantic import BaseModel,Field
class StructuredOutput(BaseModel):
    name : str = Field(description="name")
    age : int = Field(description="age")
    email : str = Field(description="email address")
```

```python
Structuredllm = llm.with_structured_output(StructuredOutput)
```

```python
response = Structuredllm.invoke(text)
print(response)
print(response.age)
print(response.name)
```

```python
print(type(response)) ##object of class StructuredOutput 
```

```python
response.model_dump()
```

```python
from typing import List
from pydantic import BaseModel, Field
class Movie(BaseModel):
    name : str = Field(description="Title of the movie")
    year : str = Field(description="Movie release year")

class AllMovies(BaseModel):
    movies : List[Movie]

movie_llm = llm.with_structured_output(AllMovies)
```

```python
response=movie_llm.invoke("give me two movies")
response.model_dump()
```

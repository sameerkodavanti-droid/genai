```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq
```

```python
load_dotenv()
```

```python
from langchain.tools import tool
```

```python
@tool
def add(a:int,b:int):
    """
    will return the sum of two numbers

        Args:
            a : Number 1
            b : Number 2
    """
    return a+b

@tool
def mul(a:int,b:int):
    """
    will return the product of two numbers

        Args:
            a : Number 1
            b : Number 2
    """
    return a*b



@tool
def div(a:int,b:int):
    """
    will return the division of two numbers

        Args:
            a : Number 1
            b : Number 2
    """
    return a//b


@tool
def sub(a:int,b:int):
    """
    will return the difference of two numbers

        Args:
            a : Number 1
            b : Number 2
    """
    return a-b
```

```python
add.invoke({"a":9,"b":5})
```

```python
from langchain.agents import create_agent
```

```python
llm = ChatGroq(
    groq_api_key=os.getenv("groq_api"),
    model="openai/gpt-oss-120b",
)
```

```python
agent = create_agent(
    model=llm,
    tools=[add,mul,sub,div],
    system_prompt="You are a math prof and always use tool for calculation,structure the output in clean readable plain Text only."

)
```

```python
res = agent.invoke({
    "messages":[
        {"role":"user","content":"what is 50*1000+102?"}
    ]
})
print(res["messages"][-1].content)
```

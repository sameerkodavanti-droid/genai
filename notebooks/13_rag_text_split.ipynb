```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text= """Friends,
If we empower our youth, strengthen our education system, and support our farmers, no force can stop us from achieving prosperity.

Let us walk this path together—towards growth, stability, and a brighter future.

Thank you. Jai Hind."""
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=20)
chunks = splitter.split_text(text)
print(chunks)
```

### document load and split

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader("../data/medical_report.pdf")
pdf = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
```

```python
pdf_chunks = splitter.split_documents(pdf)
```

```python
print(pdf_chunks)
```

##### Load ENV Variables

```python
import os
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python

load_dotenv()

groq_api = os.getenv("groq_api")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api
)
```

```python

response= llm.invoke("what is genai?")
print(response.content)
```
